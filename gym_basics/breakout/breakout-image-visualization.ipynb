```python
import warnings
warnings.filterwarnings('ignore')   

import gym
from gym.wrappers import Monitor
import numpy as np
import tensorflow as tf
import numpy as np
import sys

if "../" not in sys.path:
    sys.path.append("../")
     

import plotting
from collections import deque, namedtuple

env = gym.envs.make("Breakout-v0")

valid_actions = [0, 1, 2, 3]
```

```python
plt.imshow(state, cmap="gray")
```

```python
import tensorflow as tf


with tf.Session() as sess:
    input = tf.placeholder(shape=[210, 160, 3], dtype=tf.uint8)
    output = tf.image.rgb_to_grayscale(input)
    output = tf.image.crop_to_bounding_box(output, 34, 0, 160, 160)
    output = tf.image.resize_images(output, [84, 84], method=tf.image.ResizeMethod.NEAREST_NEIGHBOR)
    output = tf.squeeze(output)
    
    
init = tf.initialize_all_variables()
foo = sess.run(output, { input: state })    
plt.imshow(foo, cmap="gray")
```
